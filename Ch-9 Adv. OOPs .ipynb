### i) overloading
- allows programers to write methods that can process a variety of diff. types of functionality with the same name
- Method overloading is a compile type polymorphisim using which we can define 2 or more method in same class with the same name but diff. parameter list

```python
def sum_num(*args):
    r=0
    for num in args:
        r+=num
    print("Sum :",r)
print("similar the method of overloading concept")
print("1 argument ~>",end=" ")
sum_num(10)
print("2 argument ~>",end=" ")
sum_num(10,20)
print("Multiple argument ~>",end=" ")
sum_num(10,50,70,13)
```

# ii) overriding
- a child class method overrides the parent class method of the same name parameter and return type 
- it is used to overwrite or redefine a parent class  method in a derived class

```python
#Parent Class
class A:
    def first(self):
        print("first function of Class A")
    def second(self):
        print("Second function of Class A")
#Child Class
class B(A):
    def first(self):
        print("Redefine function of Class A in Class B")
    def display(self):
        print("Display function of child Class")
```

- __ add __ (self, other) for < (+)
- __ sub __ (self, other) for < (-)
- __ mul __ (self, other) for < (*)
- __ truediv __ (self, other) for < (/)
- __ floordiv __ (self, other) for < (//)
- __ mod __ (self, other) for < (%)

```python
class D:
    def __init__(self,a):
        self.a=a
    def __ne__(self,other):
        return self.a!=other.a
    def __mul__(self,other):
        return self.a*other.a
    def __truediv__(self,other):
        return self.a/other.a
    def __floordiv__(self,other):
        return self.a//other.a
    def __mod__(self,other):
        return self.a%other.a
    def __pow__(self,other):
        return self.a**other.a
obj1=D(1234)
obj2=D(7)
```

```python
if obj1 :
    print("Not Equal")
print(obj1.__mul__(obj2))
print(D.__pow__(obj1,obj2))
print(obj1.__truediv__(obj2))
print(D.__floordiv__(obj1,obj2))
```

# 2) Inhertance


- we can use a existing class to create a new class rather than creating it form scratch
- Allows the programm to reuse the code 
- with inhertance child class gain access to all data members function and properties define in the parent class
- a child class may also offers it particular implimantation of the parent class's functions 
- a derived class can inherit from its base class by simply putting its name in brackets after the derived class name 

## 1) simple inheritance



```python
class Parent:
    def fun1(self):
        print("Hello from Parent")

class Child(Parent):  # Child inherits from Parent
    def fun2(self):
        print("Hello from Child")

obj = Child()
obj.fun1()
obj.fun2()
```

## 2) Multiple inheritance

#### # Method Resolution Order (MRO)

```python
class Parent1:
    def fun1(self):
        print("Method A from Parent 1")

class Parent2:
    def fun2(self):
        print("Method B from Parent 2")

class Parent3:
    def fun(self):
        print("Method C from Parent 3")
        
class Child(Parent1, Parent2,Parent3):
    def fun3(self):
        print("Hello from Child")

# Create an object of the Child class
obj = Child()

# Access methods from both parent classes
obj.fun1()
obj.fun2()
obj.fun3()
obj.fun()
print(Child.__mro__)
```

## 3) Multi level inheritance


```python
class Grandparent:
    def fun1(self):
        print("Hello (Grandparent)")

class Parent(Grandparent):
    def fun2(self):
        print("Hiii (Parent)")

class Child(Parent):
    def fun3(self):
        print("Oyyy (Child)")

# Create an object of the final child class
obj = Child()

# Call methods from all levels of the hierarchy
obj.fun1()
obj.fun2()
obj.fun3()
```

## 4) Hierarchical  inheritance


```python
class Student:
    def fun1(self):
        print("Student")

# First Level of Hierarchy
class ART(Student):
    def get(self):
        print("Art Student")

class ENGINEERING(Student):
    def get(self):
        print("Engineering Student (General)")

class MEDICAL(Student):
    def get(self):
        print("Medical Student")

# Second Level of Hierarchy (Specific Engineering Fields)
class MECH(ENGINEERING):
    def get(self):
        print("Mechanical Engineering Student")

class ELEC(ENGINEERING):
    def get(self):
        print("Electrical Engineering Student")

class CS(ENGINEERING):
    def get(self):
        print("Computer Science Student")
        
        
obj_art = ART()
obj_art.get()  # Specific to ART

obj_eng = ENGINEERING()
obj_eng.get()  # Specific to ENGINEERING

obj_med = MEDICAL()
obj_med.get()  # Specific to MEDICAL

obj_mech = MECH()
obj_mech.get()  # Specific to MECH (Overrides ENGINEERING.get())

obj_elec = ELEC()
obj_elec.get()  # Specific to ELEC (Overrides ENGINEERING.get())

obj_cs = CS()
obj_cs.get()
obj_cs.fun1()
```

```python
class Parent:
    def __init__(self):
        self.attr1 = 50
        self.attr2 = 25

class Child(Parent):
    def __init__(self):
        # Use super() to call the Parent constructor
        super().__init__() # OR Parent.__init__(self) 
        self.attr3 = 1000

obj=Child()
print(obj.attr1)
print(obj.attr2)
print(obj.attr3)
```

###  Method Resolution Order (MRO)
- it denotes a way a programming languages resolves a method or attribute 
- mro defines the order in which the ase classes are searched when exicuting method
- first the method is searched within the class and then it follows the order we specified while inheriting 

```python
class A:
    def rk(self):
        print("Class A")
class B(A):
    def rk(self):
        print("Class B")
class C(A):
    def rk(self):
        print("Class C")
r=B()
r.rk()
print(A.__mro__)
print(B.__mro__)
print(C.__mro__)
```

```python
class A:
    def rk(self):
        print("Class A")
class B(A):
    def rk(self):
        print("Class B")
class C(A):
    def rk(self):
        print("Class C")
class D(B,C):
    def rk(self):
        print("Class D")
r=D()
r.rk()
print(A.__mro__)
print(B.__mro__)
print(C.__mro__)
print(D.__mro__)
```

```python
from abc import ABC,abstractmethod
class shape(ABC):
    def __init__(self,shapename):
        self.shapename=shapename
    @abstractmethod
    def draw(self):
        print("Drawing : ",shapename)
class circle(shape):
    def __init__(self):
        super().__init__("Circle")    
    def draw(self):
         print("Drawing :",self.shapename)
            
obj=circle()
obj.draw()
```

```python
class book:
    def __init__(self,name,noofauthor,listofother,publisher,isbn,year):
        self.name=name
        self.noofauthor=noofauthor
        self.listofother=listofother
        self.publisher=publisher
        self.isbn=isbn
        self.year=year
    def display(self):
        print("Book Name:", self.name)
        print("Number of Authors:", self.noofauthor)
        print("List of Others:", self.listofother)
        print("Publisher:", self.publisher)
        print("ISBN:", self.isbn)
        print("Year:", self.year)
class coursebook(book):
    def __init__(self,name,noofauthor,listofother,publisher,isbn,year,course_name):
        super().__init__(name,noofauthor,listofother,publisher,isbn,year)
        self.course_name=course_name
    def display(self):
        super().display()
        print("course_name:", self.course_name)
        
cb1 = coursebook("Python", 7, ["A", "B", "C"], "Tech-Press", "1234-5678-9101-1121", 2022, "Computer Science")
cb1.display()
```
